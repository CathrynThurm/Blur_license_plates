```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
license_img = cv2.imread('C:/Users/Cathryn/Programming/Python/CV_Beginning/Computer-Vision-with-Python/DATA/car_plate.jpg')
```

```python
def display_image(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    figure = plt.figure(figsize=(12,10))
    axes = figure.add_subplot(111)
    axes.imshow(img)
```

```python
display_image(license_img)
```

```python
license_cascade = cv2.CascadeClassifier('C:/Users/Cathryn/Programming/Python/CV_Beginning/Computer-Vision-with-Python/DATA/haarcascades/haarcascade_licence_plate_rus_16stages.xml')
```

```python
def detect_license(img):
    license = img.copy()
    
    rect = license_cascade.detectMultiScale(license)
    
    
    for (x,y,w,h) in rect:
        license = cv2.rectangle(license, (x,y), (x+w, y+h), (0,0,255), 3)
        
    return license
```

```python
rect_result = detect_license(license_img)
```

```python
display_image(rect_result)
```

```python
def blur_license(img):
    license = img.copy()
    roi = img.copy()
    
    rect = license_cascade.detectMultiScale(license)
    
    
    for (x,y,w,h) in rect:
        roi = roi[y:y+h, x:x+w]
        blurred_roi = cv2.medianBlur(roi, 7)
        license[y:y+h, x:x+w] = blurred_roi
        
    return license
```

```python
blur_result = blur_license(license_img)
```

```python
display_image(blur_result)
```
